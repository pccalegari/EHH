**Exemplo Método de Euler**

 Considere o problema,

  \begin{equation}
    \left\{\begin{array}{ll}
    y(0) & = 0, t\in [0,1]\\
    \dfrac{dy}{dt} &= y^2(t) + 2y(t) - t^4,
    \end{array}\right .    
\end{equation}

cuja solução exata é $y(t)=t^2.$  

```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    #Condições iniciais
    t0 = 0
    y0 = 0
    tn = 1
    #Discretização do intervalo
    n = 5
    dt = (tn - t0)/n
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    ye = np.zeros(n+1)
    e = np.zeros(n+1)
    #Inicialização
    t[0] = t0
    y[0] = y0
    ye[0] = sol(t0)
    e[0] = abs(ye[0] - y[0])

    for k in range(n):
        t[k+1] = t[k] + dt
        y[k+1] = y[k] + dt*f(t[k], y[k])
        ye[k+1] = sol(t[k+1])
        e[k+1] = abs(y[k+1] - ye[k+1])
        #Comente a linha abaixo se n for grande
        print(t[k+1], y[k+1], ye[k+1], e[k+1])

    print(n,t[n], e[n])

    plt.figure(1)
    plt.plot(t, y, 'r*', label = 'Sol. Apr.')
    plt.plot(t, ye, 'b-.', label = 'Sol. Exa.')
    plt.legend()
    plt.xlabel('t')
    plt.ylabel('y')
    plt.show()

    plt.figure(2)
    plt.plot(t, e, 'r-.')
    plt.xlabel('t')
    plt.ylabel('Erro')
    plt.show()

def f(t,y):
    return y*y + 2*t - t**4

def sol(t):
    return t*t

main()
```
